##### Chapter 4....data analysis and visualization


```python
#Which year had the best car prices
```

```python
results = df.groupby('year').sum()
years= range(2010,2019)
import matplotlib.pyplot as plt
plt.bar(years,results['price'])
plt.xticks(years)
plt.xlabel('YEARS')
plt.ylabel('PRICES IN US DOLLORS(IN TENS OF THOUSANDS)')
plt.title('CAR PRICES FROM 2010-2018')
plt.show()
```

```python
#which model sold the most cars
```

```python
model_sales=df.groupby("model")['price'].sum().nlargest(9)
model_sales
import seaborn as sns
sns.barplot(y=model_sales.index, x=model_sales).set(title='TOP 5 MODELS WHICH SOLD MOST')
```

```python
model_sales=df.groupby("model")['price'].sum().nlargest(9)
model_sales
```

```python
#which model has the lowest car mileage?
```

```python
model_mileage=df.groupby("model")['mileage'].sum().nlargest(9)
model_mileage
import seaborn as sns
sns.barplot(y=model_mileage.index, x=model_mileage).set(title='TOP 5 MODELS WITH THE LOWEST MILEAGE')
```

```python
model_mileage=df.groupby("model")['mileage'].sum().nlargest(9)
model_mileage
```
